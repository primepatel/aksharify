```python
from aksharify import TextArt

# input image file
art = TextArt("sample5.png") # http://paulbourke.net/fractals/juliaset/
```

```python
# set dimension of output text
art.set_dim(160, 40)

# genetate text
art.asciify()

# # output into .txt file
# art.text_output("output")

# generate colored text using html
art.svgify()

# # output into .html file
# art.html_output("output")

# generate colored text using svg
art.svgify()

# output into .svg file
art.svg_output("output")

```
